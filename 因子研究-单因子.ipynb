```python
import re
import math
df = DataSource('bar1d_CN_FUTURE').read() # 读取期货日线数据 
df['exchange'] = df['instrument'].apply(lambda x:x.split('.')[-1]) # 交易所
df['contract'] = df['instrument'].apply(lambda x:x.split('.')[0]) # 合约代码
df['future_name'] = df['contract'].apply(lambda x:re.sub('[^a-zA-Z]', '', x)) # 期货品种
df['contract_info'] = df['contract'].apply(lambda x:x[-4:])
continus_contract_df = df[df['contract_info'] =='8888'] # 将主力合约数据过滤出来

rabalance_period = 22 # 调仓天数
long_port_is_big_factor = True   # 方向
buy_commission_rate = 0.0003 # 买入佣金
stamp_tax = 0.001 #税
sell_commission_rate = 0.0003 # 卖出佣金

def handle_data(df):
    # 计算当期因子和未来一段时间收益率
    df['factor'] = df['close'] / df['close'].shift(20) - 1 # 
    df['ret'] = df['close'].shift(-1*rabalance_period) / df['close'] - 1
    return df
factor_df = continus_contract_df.groupby('contract').apply(handle_data)

start_date = factor_df['date'].min()
daily_factor_df = factor_df[['date','close','factor','ret','contract']]

# 只取出调仓日的数据
td = D.trading_days(start_date=start_date)
rebalance_days = td[::rabalance_period] # 调仓期 
rebalance_days_df = pd.DataFrame({'date':rebalance_days['date'] , 'ix':range(len(rebalance_days))})
rebalance_days_df.index = range(len(rebalance_days_df))

merge_df = pd.merge(daily_factor_df, rebalance_days_df, on='date', how='inner')
merge_df = merge_df.drop('ix', axis=1)

def cal_ic(df):
    # 计算IC 
    return df[['factor','ret']].corr().values[0][1]

ic = merge_df.groupby('date').apply(cal_ic)
ic_mean = ic.mean()
ic_std = ic.std() 

# 画IC图
T.plot(ic, chart_type='column', title='IC值')


# 分组
def cut_box(tmp):
    if tmp.factor.isnull().sum() == len(tmp): # 因子值全是nan的话
        tmp['factor_group'] = [np.nan]*len(tmp)
    else:
        tmp['factor_group'] = pd.qcut(tmp['factor'], 5, labels=range(5)) # 升序排序，分成5组
    return tmp 

group_df  = merge_df.groupby('date').apply(cut_box)
result = group_df[['date', 'factor_group', 'ret']].groupby(['factor_group', 'date']).mean().reset_index()

result['ret'] -= (buy_commission_rate+sell_commission_rate+stamp_tax)
result_table = result.pivot(values='ret', columns='factor_group', index='date')
result_table.rename(columns={0:'top1_ret',1:'top2_ret',2:'top3_ret',3:'top4_ret',4:'top5_ret'}, inplace=True)

# 合并基准数据
market_index_code = '000300.HIX'
start_date  = result_table.index.min().strftime('%Y-%m-%d')
end_date  = result_table.index.max().strftime('%Y-%m-%d')

bm_code = market_index_code
bm_df = DataSource('bar1d_index_CN_STOCK_A').read(start_date=start_date, end_date=end_date, instruments=[bm_code], fields=['close'])
bm_df['bm_ret'] = bm_df['close'].pct_change().fillna(0)
result_table = pd.merge(result_table.reset_index(), bm_df[['date', 'bm_ret']], on='date', how='inner').fillna(0).set_index('date')
if long_port_is_big_factor:
    result_table['LS_ret'] = result_table['top5_ret'] - result_table['top1_ret'] 
else:
    result_table['LS_ret'] = result_table['top1_ret'] - result_table['top5_ret'] 

for i in result_table.columns: 
    col_name = i.split('_')[0]+ '_' + 'pv' 
    relative_ret_name = i.split('_')[0]+ '_' + 'relative_pv'
    result_table[col_name] = (1+result_table[i]).cumprod()
    result_table[relative_ret_name] = (1+(result_table[i]-result_table['bm_ret'])).cumprod()

if long_port_is_big_factor:
    result_table['long_bm_ret'] = result_table['top5_ret'] - result_table['bm_ret']
else:
    result_table['long_bm_ret'] = result_table['top1_ret'] - result_table['bm_ret']
    
# 绘制分组收益率
T.plot(result_table[['top1_pv', 'top2_pv','top3_pv', 'top4_pv', 'top5_pv']])
```
